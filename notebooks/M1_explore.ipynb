```python
import pandas as pd
import sqlalchemy
import sqlite3
```

```python
enn = sqlalchemy.create_engine('sqlite:///../data/raw/raw_data_project_m1.db')

personal = pd.read_sql_table(table_name='personal_info', con=enn)

personal
```

```python
personal.info()
```

```python
personal.describe()
```

```python
personal.rename(index={e: e+1 for e in range(len(personal))}, 
                columns={'dem_has_children': 'children'}, inplace=True)
personal.head()
```

```python
import re

personal['age'] =  personal['age'].apply(lambda x: re.sub('\D', '', x)).astype(int)
```

```python
from datetime import datetime

def agefunc(x):
    today = datetime.today().year
    y = 0
    if x <= 1000:
        pass
    else:
        y = today - x
        return y
```

```python
m = personal['age'] <= 1000
personal['age'] = personal['age'].where(m, personal['age'].apply(agefunc))
personal
```

```python
personal['gender'] = personal['gender'].str.lower().apply(lambda x: re.sub('fem$', 'female', x))
personal
```

```python
personal['children'] = personal['children'].str.lower()
personal.head()
```

```python
#Assigning 'juvenile' to an age_agroup.

personal['age_group'].value_counts()
```
